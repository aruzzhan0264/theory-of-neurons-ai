```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

create syntetic dataset

```python
def f(x_data):
    y_data= x_data ** 2 + 1
    return y_data
x0=np.array([np.linspace(0,10,47)])
x=x0.T
print('x.shape=',x.shape)
print('x0.shape=',x0.shape)
y=f(x)
#y1=y[:,0]
#y1.shape
plt.figure(figsize=(3,4))
plt.plot(x,y,'.b')
```

Gradient descent preparation

```python
m=x.size
print('m=',m)
ones=np.ones([m,1])
print('ones=',ones)
X=np.concatenate((ones,x),axis=1)
X
```

```python
theta=np.array([50.5, 0.71])
print('theta=',theta)
h=np.dot(X,theta)
h
```

```python

plt.figure(figsize=(8,10))
plt.plot(x,y,'xb')
plt.plot(x,h,'-y')
alpha=0.25
iterations=100
```

Gradient descent loop

```python
alpha=0.025
iterations=1000

plt.figure(figsize=(8,10))
plt.plot(x,h,'-y')
plt.plot(x,y,'xb')
for i in range(iterations):
    theta[0]=theta[0]- alpha*(1/m)*np.sum((h-y[:,0])*X[:,0])
    theta[1]=theta[1]- alpha*(1/m)*np.sum((h-y[:,0])*X[:,1])
    h=np.dot(X,theta)
plt.plot(x,h,'-r')    
#None
#print('sse=',sse)
```

```python
# 1.     theta
start_time = time.time()
theta_matrix = np.linalg.inv(X.T @ X) @ X.T @ y
end_time = time.time()

print("Theta   :", theta_matrix)
print(",    theta:", end_time - start_time)
```

```python
# 2.  MSE
h_matrix = np.dot(X, theta_matrix)
mse_matrix = np.mean((h_matrix - y) ** 2)
print("  (MSE)   :", mse_matrix)
```

```python
# 3.  MSE   
mse_gradient = np.mean((h - y) ** 2)
print("  (MSE)   :", mse_gradient)
```
